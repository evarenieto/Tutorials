```python
# creating empty graph and adding nodes
import networkx as nx

G=nx.Graph()

# adding just one node:
G.add_node("a")
# a list of nodes:
G.add_nodes_from(["b","c"])

print("Nodes of graph: ")
print(G.nodes())
print("\nEdges of graph: ")
print(G.edges())
```

```python
#adding edges
G.add_edge(1,2)
edge = ("d", "e")
G.add_edge(*edge)
edge = ("a", "b")
G.add_edge(*edge)

print("Nodes of graph: ")
print(G.nodes())
print("\nEdges of graph: ")
print(G.edges())
```

```python
# adding edges as a list

G.add_edges_from([("a","c"),("c","d"), ("a",1), (1,"d"), ("a",2)])
```

```python
print("Nodes of graph: ")
print(G.nodes())
print("\nEdges of graph: ")
print(G.edges())
```

```python
# ploting time :) and saving for report
%matplotlib notebook
import matplotlib.pyplot as plt
nx.draw_networkx(G)
plt.savefig("simple_path.png") # save as png
plt.show() 
```

```python
Gr=nx.path_graph(4)
%matplotlib inline
print("Nodes of graph: ")
print(Gr.nodes())
print("Edges of graph: ")
print(Gr.edges())
nx.draw_networkx(Gr)
plt.savefig("path_graph1.png")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

GRR=nx.path_graph(4)
cities = {0:"Madrid",1:"London",2:"Barcelona",3:"New York"}

H=nx.relabel_nodes(GRR,cities)
 
print("Nodes of graph: ")
print(H.nodes())
print("Edges of graph: ")
print(H.edges())
nx.draw_networkx(H) #with_labels=False
plt.savefig("path_graph_cities.png")
plt.show()
```
